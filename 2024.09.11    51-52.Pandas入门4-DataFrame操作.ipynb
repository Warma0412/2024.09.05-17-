```python
import pandas as pd
```

```python
s_name = pd.Series(["沃玛", "憨色", "冷鸟", "吴优"], index=["001", "002", "003", "004"])
s_sex = pd.Series(["女", "女", "女", "男"], index=["001", "002", "003", "004"])
s_height = pd.Series([160.0, 158.5, 162.2, 172.5], index=["001", "002", "003", "004"])
s_grade = pd.Series([100, 90, 80, 66], index=["001", "002", "003", "004"])

df_1 = pd.DataFrame({"姓名":s_name, "性别":s_sex, "身高":s_height, "成绩":s_grade})
df_1
```

```python
df_1["成绩"]    # 获取一列数据
```

```python
s_grade_new = pd.Series([100, 60, 60, 60], index=["001", "002", "003", "004"])

df_1["成绩"] = s_grade_new    # 在 获取一列数据 命令的基础上增加 = 新的Series即可
df_1                         # 注意：如果DataFrame有标签索引，更换的Series也要带有标签索引才能正确识别
```

```python
df_1["成绩"] = [100, 70, 70, 60]    # 替换数据使用python列表也是可以的，无需标签索引，只要列表长度与DataFrame吻合便可自动根据顺序对齐
df_1
```

```python
df_1["班级"] = ["一班", "二班", "二班", "一班"]    # 赋值时若列名不存在，则会根据列名与值创建一个新的列 --- 对列赋值即可以用来更新列，也可以用来创建列
df_1
```

**（2）更新DataFrame中的行:**

```python
df_1.loc["004"] =pd.Series(["桐人", "男", 172.0, 90, "二班"], index=["姓名", "性别", "身高", "成绩", "班级"])
df_1
```

**（3）删除DataFrame中的行or列:**

        删除行：df.drop(["002", "004"])----注意：.drop()方法会自动返回删除后的结果，但不会修改原DataFrame，真删则赋值df_2 = df.drop()
        删除列：df.drop(["身高", "性别"], axis=1)----axis=1即进行横向操作，即沿着横向方向对列进行操作（默认axis=0，即对行进行操作）

```python
df_1.drop(["002", "004"])     # 在不改变原DataFrame的基础上，自动返回删除后的结果
```

```python
df_2 = df_1.drop(["身高", "班级"], axis=1)    # 对列进行drop操作一定要设置 axis =1（更多的是对于行，即实例的删减，默认axis=0更实用）
df_2
```

**（4）DataFrame与DataFrame之间的操作:**

        正如Series加Series的时候，会自动根据索引进行对齐；
        当DataFrame加Dataframe时，索引和索引、列名和列名也会自动对齐；

        如果是DataFrame与DataFrame之间操作，则可用.add()、.sub()、.mul()、.div()

```python
s_a = pd.Series([1, 0, 2, 3], index=["001", "003", "004", "005"])
s_b = pd.Series([4, 2, 3, 5], index=["001", "003", "004", "005"])
s_c = pd.Series([160.0, 158.5, 162.2, 172.5], index=["001", "003", "004", "005"])
s_d = pd.Series([3, 2, 1, 0], index=["001", "003", "004", "005"])

df_3 = pd.DataFrame({"a":s_a, "b":s_b, "c":s_c, "d":s_d})
df_3
```

```python
s_a = pd.Series([0, 0, 1], index=["002", "004", "005"])
s_b = pd.Series([2, 0, 1], index=["002", "004", "005"])
s_c = pd.Series([30.0, 12.5, 13.2], index=["002", "004", "005"])

df_4 = pd.DataFrame({"a":s_a, "b":s_b, "c":s_c})
df_4
```

```python
df_3 + df_4    # 这里004和005在都存在于两个DataFrame中，便将两者共有的abc三列的数据进行了相加----且将只存在于df_4中的002也添加了进去
```

```python
df_3.add(df_4)    # 与上面的+法一样，由于其他两个DataFrame存在各自独有的数据，在相加时会默认为NaN，即无数据，返回的结果也会是NaN
```

```python
df_3.add(df_4, fill_value=0)    # 使用fill_value：df_3中的存在的元素，对应在df_4中找不到的话，便会默认为0
                                # 但某个元素值在两个DataFrame中都不存在，即压根没有进行相加操作，因此也不存在替换，还是NaN----详细可见002的d列
```

        同样也可以使用.sub()、.mul()、.div()----减、乘、除

**（5）DataFrame与Series之间的操作:**

        此时是用Series的索引与DataFrame的列名进行对齐,然后把操作运用在DataFrame的每一行上----对齐不了的是NaN（文字不太好理解）

```python
s_mul = pd.Series([1, 2, 0.01, 3],index = ["a", "b", "c", "e"])
s_mul
```

```python
df_3
```

```python
df_3 * s_mul    # s_mul中索引a的值，会与df_3中a列下面的每一个值进行相乘----又体现了广播机制（不同维度之间的运算机制）
                # s_mul中索引d的值是NaN，相乘后使得原本有数据的d列也变成了NaN
                # s_mul有索引e的值，但df_3中没有e列的值，也是NaN，即相乘后的结果也是NaN
```
