#### 一、JSON文件

**（1）文件后缀介绍:**

        文件后缀名只是文件名的一部分， 更改后缀不影响实际的文件格式（内容）
                后缀名会影响系统使用什么软件打开它，不影响内容----如系统会用音乐播放器打开.mp3

**（2）JSON数据格式:**

        全称JavaScript Object Notation
        JSON也有两种数据结构：1.JSON对象：python字典{}
                           2.JSON数组：python列表[]

        一个JSON对象:实质是一个dict

```python
{
 "name":"warma",
 "sex":"femal",
 "age":"26"  
}
```

        一个JSON数组:实质是一个list，但其中的每个元素都是dict

```python
[
 {
 "name":"warma",
 "sex":"femal",
 "age":"26"  
 },
 {
 "name":"hanser",
 "sex":"femal",
 "age":"31"  
 }
]
```

**（3）JSON数据格式·规则:**

        额外规则：1.python字典中允许让int数据作为key，但json数据中只能是字符串作为key，且该字符串必须被双引号""包围，''不行
                2.json中的值必须是 字符串、数字、布尔值true、数组[]、对象{}、空值null 中的一种 ---- 布尔值遵循java规则，为小写字母开头
                3.key不能重复
                4.对于第二点，JSON数据类型可以是第二点中的六种数据的任意搭配

        在确定文件是JSON类型后，就可以直接用python进行分析了吗？
                NO，因为json与python的字典与列表或多或少还存在差别，如bool----必须进行解析才能进行下一步分析（利用pandas解析+转化）

#### 三、CSV文件

**（1）CSV数据格式:**

        全称Comma-Saperated Values，表示逗号分隔值
        CSV与JSON一样是纯文本格式，其结构基本是一个表格，在excel等表格软件打开便可以看到清晰表示
                一般文件第一行是表格头，即展示列名的那一行（也可以没有标题头）
                表格头下面，每一行数据独占一行----每一行的逗号分隔符的数量一致，否则就不是合格的CSV文件了 

===============================================================================

        CSV是一个非常规整的二维结构，与DataFrame一致，基本上一眼可以看出对应表格格式
        JSON则不同，它的结构非常灵活，也可以层层嵌套，难以看出对应表格格式

        word、excel等文件不是纯文本格式，因此会编程的人使用代码进行数据分析时，尽量会让源数据以CSV、JSON等纯文本格式、或者sqlite等数据库格式保存

**（2）打开数据注意事项:**

        如果一个CSV文件大小1G左右，说明数据量十分庞大，若用xcel打开可能导致崩溃，可在纯文本编辑器中打开，更好的是转化为DataFrame选择性查看部分数据

#### 四、读取数据-CSV

**（1）pd.read_csv函数:**

        CSV格式的优点：体积小、结构工整、易理解、可直接用excel打开

```python
pd.read_csv("/Users/warma/Downloads/fifa_players_2022.csv")   
```

        注意：以上数据有表格头，会自动将第一行看作列名，如果第一行直接就是数据，则添加header=None，如pd.read_csv("文件路径", header=None) 
             第一行会被默认解析为列名，但不会讲第一列解析为索引，而是默认从0开始的位置索引
                     如果想将数据中的其中一列设定为标签索引时，则添加index_col="列名"，如pd.read_csv("文件路径", index_col="sofifa_id") 

```python
pd.read_csv("/Users/warma/Downloads/fifa_players_2022.csv", index_col="sofifa_id") 
```

        当数据不仅特别长，还特别宽时，即列名过多，则部分列名会被隐藏在···中，若想展示全部的列名，则可使用pd.set_option()函数

```python
pd.set_option("display.max_columns", 150)    # 将展示列数上限调整为150（默认为20列）
```

```python
pd.read_csv("/Users/warma/Downloads/fifa_players_2022.csv") 
```

        当部分列的数据内容过长，无法完全展示，如player_url，也可以调用pd.set_option()函数

```python
pd.set_option("display.max_colwidth", 500)    # 将内容展示字符上限调整为500（默认为50）
```

```python
pd.read_csv("/Users/warma/Downloads/fifa_players_2022.csv") 
```
